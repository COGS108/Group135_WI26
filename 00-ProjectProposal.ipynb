## Research Question

Does the wingspan of an NBA player affect their overall performance? We will consider the wingpspan of a player and their respective advanced metrics, such as RAPTOR, LEBRON, EPM, DARKO, Net Rating and True Shooting Percentage. 

## Background and Prior Work

Most humans have about a 1:1 height to wingspan ratio. However, when it comes to athletic endeavors, some combination of features can lead to greater outcomes, for example Michael Phelps is frequently cited as being anatomically perfect to perform in the sport of swimming. We plan to examine wingspan in NBA players to see if players with longer wingspans tend to have better outcomes in the NBA. Better outcomes will be quantified using Robust Algorithm using Player Tracking and On/Off Ratings (RAPTOR), estimated plus-minus (EPM), luck-adjusted player estimate using a box prior regularized on-off (LEBRON), daily adjusted and regressed kalman optimized (DARKO), net rating, and true shooting. 
RAPTOR represents the overall influence on the match that a player has depending on when they are on or off the court.
EPM further elicits the individual impact that a player has by splitting offensive and defensive value and compensating for outside influences.
LEBRON the random environmental chance that affects the offensive and defensive impact of a player on the match into account and also relative to their positions. 
DARKO helps to normalize the history of a player and places slightly more weight on their recent statistics. 
Net rating calculates the differences between the offensive and defensive rating, finding the primary type of impact that a player has on a match.
True Shooting: an advanced NBA statistic measuring scoring efficiency by accounting for 2-point field goals, 3-point field goals, and free throws in one metric.

We found two studies that analyze wingspan in basketball. The first study, “Allometric Variation in Modern Humans and the Relationship Between Body Proportions and Elite Athletic Success”, suggests that there are benefits to having longer wingspans. In the second study, “Basketball Players’ Long Wingspan Enhances Defensive Upside but Hinders Shooting Capabilities”, the researcher comes to the conclusion that long wingspans contribute to defensive capabilities, but hinder shooting capabilities.

## Hypothesis


Our main hypothesis about the answer to our question is that there may be a positive correlation between player wingspan and a player’s overall performance. The reason for this is that with a longer wingspan, a player may be able to have a wider area of influence which allows them to perform more effectively
Our current understanding is that usually longer wingspans positively affect defensive value, as it is easier to block shots and deflect the ball with a longer reach, and that people with wider bodies tend to be stronger. (longer arms/wider bodies). We're also not sure how exactly wingspan affects offensive potential, as it seems that there are more unique outcomes for what leads to good offense. For now, we will say that, because of the prior work we have read, it may affect offense negatively because it affects shooting negatively. In terms of overall performance, the Defensive benefits seem to surpass the offense negatives, so there will be a net positive from having a longer wingspan.
