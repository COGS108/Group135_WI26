```python
wingspans=pd.read_csv("01-interim/wingspan_cleaned/combined_cleaned_true_final.csv")
wingspans.head(3)
```

```python
def normalize_name(name):
    if "," in name:
        parts=name.split(",", 1)
        last=parts[0].strip()
        first=parts[1].strip()
        return f"{first} {last}"
    else:
        return name.strip()
```

```python
wingspans=wingspans.drop(columns=["Unnamed: 0"])
```

```python
wingspans=wingspans.rename(columns={"name":"Player"})
wingspans["Player"]=wingspans["Player"].apply(normalize_name)
wingspans["relative"]=wingspans["relative"].round(2)
wingspans=wingspans.drop_duplicates(subset=["Player"])
wingspans
```

```python
final_df=pd.concat([final_2022, final_2021, final_2020, final_2019, final_2018])
final_df
```

```python
final_df['Player'] = final_df['Player'].str.replace(r'\s+', ' ', regex=True).str.strip()
wingspans['Player'] = wingspans['Player'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
true_final = final_df.merge(
    wingspans[["Player", "height_inches","wingspan_inches", "relative"]],
    on="Player",
    how="left"
)
true_final=true_final.dropna()
true_final.to_csv("02-processed/final_data.csv", index=False)
true_final
```
