```python
import pandas as pd
```

```python
lebron_2022=pd.read_csv("01-interim/lebron_cleaned/lebron_2022_cleaned.csv")
lebron_2021=pd.read_csv("01-interim/lebron_cleaned/lebron_2021_cleaned.csv")
lebron_2020=pd.read_csv("01-interim/lebron_cleaned/lebron_2020_cleaned.csv")
lebron_2019=pd.read_csv("01-interim/lebron_cleaned/lebron_2019_cleaned.csv")
lebron_2018=pd.read_csv("01-interim/lebron_cleaned/lebron_2018_cleaned.csv")
```

```python
raptor_2022=pd.read_csv("01-interim/raptor_cleaned/raptor_2022.csv")
raptor_2021=pd.read_csv("01-interim/raptor_cleaned/raptor_2021.csv")
raptor_2020=pd.read_csv("01-interim/raptor_cleaned/raptor_2020.csv")
raptor_2019=pd.read_csv("01-interim/raptor_cleaned/raptor_2019.csv")
raptor_2018=pd.read_csv("01-interim/raptor_cleaned/raptor_2018.csv")
```

```python
darko_2022=pd.read_csv("01-interim/darko_cleaned/darko_2022.csv")
darko_2021=pd.read_csv("01-interim/darko_cleaned/darko_2021.csv")
darko_2020=pd.read_csv("01-interim/darko_cleaned/darko_2020.csv")
darko_2019=pd.read_csv("01-interim/darko_cleaned/darko_2019.csv")
darko_2018=pd.read_csv("01-interim/darko_cleaned/darko_2018.csv")
```

```python
ts_2022=pd.read_csv("01-interim/TS_cleaned/ts_2022_clean.csv")
ts_2021=pd.read_csv("01-interim/TS_cleaned/ts_2021_clean.csv")
ts_2020=pd.read_csv("01-interim/TS_cleaned/ts_2020_clean.csv")
ts_2019=pd.read_csv("01-interim/TS_cleaned/ts_2019_clean.csv")
ts_2018=pd.read_csv("01-interim/TS_cleaned/ts_2018_clean.csv")
```

```python
on_off_2022=pd.read_csv("01-interim/on_off_cleaned/of_2022_clean.csv")
on_off_2021=pd.read_csv("01-interim/on_off_cleaned/of_2021_clean.csv")
on_off_2020=pd.read_csv("01-interim/on_off_cleaned/of_2020_clean.csv")
on_off_2019=pd.read_csv("01-interim/on_off_cleaned/of_2019_clean.csv")
on_off_2018=pd.read_csv("01-interim/on_off_cleaned/of_2018_clean.csv")
```

```python
merged_2022=pd.merge(
    darko_2022,
    raptor_2022,
    on="Player",
)
merged_2022_2=pd.merge(
    on_off_2022,
    ts_2022,
    on="Player"
)
merged_2022_3=pd.merge(
    merged_2022_2,
    lebron_2022,
    on="Player"
)
final_2022=pd.merge(
    merged_2022_3,
    merged_2022,
    on="Player"
)
final_2022["Season"]=2022
final_2022
```

```python
merged_2021=pd.merge(
    darko_2021,
    raptor_2021,
    on="Player",
)
merged_2021_2=pd.merge(
    on_off_2021,
    ts_2021,
    on="Player"
)
merged_2021_3=pd.merge(
    merged_2021_2,
    lebron_2021,
    on="Player"
)
final_2021=pd.merge(
    merged_2021_3,
    merged_2021,
    on="Player"
)
final_2021["Season"]=2021
final_2021
```

```python
merged_2020=pd.merge(
    darko_2020,
    raptor_2020,
    on="Player",
)
merged_2020_2=pd.merge(
    on_off_2020,
    ts_2020,
    on="Player"
)
merged_2020_3=pd.merge(
    merged_2020_2,
    lebron_2020,
    on="Player"
)
final_2020=pd.merge(
    merged_2020_3,
    merged_2020,
    on="Player"
)
final_2020["Season"]=2020
final_2020
```

```python
merged_2019=pd.merge(
    darko_2019,
    raptor_2019,
    on="Player",
)
merged_2019_2=pd.merge(
    on_off_2019,
    ts_2019,
    on="Player"
)
merged_2019_3=pd.merge(
    merged_2019_2,
    lebron_2019,
    on="Player"
)
final_2019=pd.merge(
    merged_2019_3,
    merged_2019,
    on="Player"
)
final_2019["Season"]=2019
final_2019
```

```python
merged_2018=pd.merge(
    darko_2018,
    raptor_2018,
    on="Player",
)
merged_2018_2=pd.merge(
    on_off_2018,
    ts_2018,
    on="Player"
)
merged_2018_3=pd.merge(
    merged_2018_2,
    lebron_2018,
    on="Player"
)
final_2018=pd.merge(
    merged_2018_3,
    merged_2018,
    on="Player"
)
final_2018["Season"]=2018
final_2018
```

```python
wingspans=pd.read_csv("01-interim/wingspan_cleaned/combined_cleaned.csv")
wingspans.head(3)
```

```python
wingspans=wingspans.rename(columns={"name":"Player"})
wingspans
```

```python
final_df=pd.concat([final_2022, final_2021, final_2020, final_2019, final_2018])
final_df
```

```python
final_df['Player'] = final_df['Player'].str.replace(r'\s+', ' ', regex=True).str.strip()
wingspans['Player'] = wingspans['Player'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
true_final = final_df.merge(
    wingspans[["Player", "height_inches","wingspan_inches", "relative"]],
    on="Player",
    how="left"
)
true_final=true_final.dropna()
true_final.to_csv("02-processed/final_data.csv", index=False)
true_final
```
