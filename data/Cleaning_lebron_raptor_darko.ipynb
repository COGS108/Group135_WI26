```python
import pandas as pd
import requests
import json
import unicodedata
```

```python
raptor=pd.read_json("00-raw/raptor_raw.json")
print(raptor.isna().sum())
raptor.head()
```

```python
darko=pd.read_json("00-raw/darko_raw.json")
print(darko.isna().sum())
darko.head()
```

```python
darko=darko.rename(columns={"player_name":"Player"})
raptor=raptor.rename(columns={"player_name":"Player"})
```

```python
darko_order=["Player", "dpm", "d_dpm", "o_dpm"]
raptor_order=["Player","raptor_total","raptor_defense","raptor_offense"]
lebron_order=["Player","LEBRON", "D-LEBRON", "O-LEBRON"]
```

```python
def remove_accents(name):
    if(isinstance(name, str)):
        return unicodedata.normalize('NFKD', name).encode('ascii', errors='ignore').decode('utf-8')
    else: return name
```

```python
darko_2022=darko[darko["season"]== 2022]
darko_2022=darko_2022[darko_order].reset_index(drop=True)
darko_2022=darko_2022.sort_values(by="dpm", ascending=False).reset_index(drop=True)
darko_2022["Player"]=darko_2022["Player"].apply(remove_accents)
darko_2022["Player"]=darko_2022["Player"].str.lower()
darko_2022=darko_2022.round({"dpm":2, "d_dpm":2, "o_dpm":2})
darko_2022.to_csv("01-interim/darko_cleaned/darko_2022.csv",index=False)
darko_2022.head()
```

```python
raptor_2022=raptor[raptor["season"] == 2022]
raptor_2022=raptor_2022[raptor_order].reset_index(drop=True)
raptor_2022=raptor_2022.sort_values(by="raptor_total", ascending=False).reset_index(drop=True)
raptor_2022["Player"]=raptor_2022["Player"].apply(remove_accents)
raptor_2022["Player"]=raptor_2022["Player"].str.lower()
raptor_2022=raptor_2022.round({"raptor_total":2, "raptor_defense":2, "raptor_offense":2})
raptor_2022.to_csv("01-interim/raptor_cleaned/raptor_2022.csv",index=False)
raptor_2022.head()
```
